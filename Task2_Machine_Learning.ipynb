**Determining the Optimal Number of Clusters**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans # Import the KMeans class

# Load the dataset
data = pd.read_csv('Mall_Customers.csv')

# Display the first few rows
data.head()

# Check for missing values and data types
data.info()

# Select features
features = data[['Annual Income (k$)', 'Spending Score (1-100)']]

# Find the optimal number of clusters
wcss = []  # Within-cluster sum of squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300)
```

**Applying** K-means Clustering

```python
# Apply K-means with the optimal number of clusters (e.g., 5)
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters = kmeans.fit_predict(features)

# Add cluster labels to the data
data['Cluster'] = clusters

# Plot the clusters
plt.scatter(features['Annual Income (k$)'], features['Spending Score (1-100)'], c=clusters, cmap='rainbow')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Clusters')
plt.show()
```
